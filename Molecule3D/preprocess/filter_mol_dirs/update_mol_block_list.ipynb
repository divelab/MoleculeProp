It seems that redirecting error message works only in jupyter notebook (in another process)

http://rdkit.blogspot.com/2016/03/capturing-error-information.html

```python
from io import StringIO
import sys
from rdkit import Chem
import os.path as osp
from tqdm import tqdm
import pickle
```

```python
with open('/mnt/dive/shared/kaleb/Datasets/PubChemQC/utils/initial_list_of_valid_mol_dirs.pkl', 'rb') as f:
    valid_dirs = pickle.load(f)

raw_dir = '/mnt/dive/shared/kaleb/Datasets/PubChemQC/raw/initial_sdfs'
sdf_paths = [osp.join(raw_dir, 'combined_mols_0_to_1000000.sdf'),
             osp.join(raw_dir, 'combined_mols_1000000_to_2000000.sdf'),
             osp.join(raw_dir, 'combined_mols_2000000_to_3000000.sdf'),
             osp.join(raw_dir, 'combined_mols_3000000_to_3982396.sdf')]

block_list_patch = {'error':[], 'warning':[]}

Chem.WrapLogs()
sio = sys.stderr = StringIO()

for sdf_path, offset in zip(sdf_paths, [0, 1000000, 2000000, 3000000]):
    print('Filtering', sdf_path)
    suppl = Chem.SDMolSupplier(sdf_path, removeHs=False, sanitize=False)
    for i, mol in enumerate(suppl):
        try:
            Chem.MolToSmiles(mol)
        except Exception as e:
            block_list_patch['error'].append((valid_dirs[i+offset], i+offset))  # (path, init_sdf_index)
            print(i + offset)
            print(f'{e}'[:-1]) # remove '\n'

        msg = sio.getvalue()
        if msg:
            print(i + offset)
            print(msg[:-1])  # remove '\n'
            if msg.startswith('RDKit WARNING'):
                block_list_patch['warning'].append((valid_dirs[i+offset], i+offset))  # (path, init_sdf_index)
                
        sio = sys.stderr = StringIO() # reset the error logger
            
save_path = '/mnt/dive/shared/kaleb/Datasets/PubChemQC/utils/mol_block_list_patch.pkl'
with open(save_path, 'wb') as f:
    pickle.dump(block_list_patch, f)
```
